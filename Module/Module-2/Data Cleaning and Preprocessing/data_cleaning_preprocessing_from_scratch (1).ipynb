## 1. What is Data Cleaning?

**Data cleaning** means finding and fixing problems in data so it becomes usable for analysis, dashboards, reporting, or machine learning.

Raw data from the real world is often messy. It may contain:

- missing values
- duplicate rows
- wrong spellings
- inconsistent formats
- wrong data types
- impossible or extreme values

If we do not clean the data, our analysis can become wrong.


## 2. Why Raw Data is Rarely Usable

Real-world data comes from many sources:

- manual entry by humans
- Excel files
- forms
- mobile apps
- APIs
- databases
- scraped websites

Because of that, data usually contains mistakes.

### Example problems
- A person's age may be written as `"25"`, `"twenty five"`, or `250`
- Gender may appear as `"Male"`, `"male"`, `"M"`, `"m"`
- A date may appear as `2025-01-05`, `05/01/2025`, or `Jan 5 2025`
- Salary may be stored as text like `"50,000"` instead of a number
- The same customer may appear multiple times

So before analysis, we clean the dataset.


## 3. Structured vs Unstructured Data

### Structured Data
Structured data has a clear table-like format.

Examples:
- Excel sheet
- SQL table
- CSV file
- sales records
- student marks table

Example:

| name | age | city |
|---|---:|---|
| Ram | 23 | Kathmandu |
| Sita | 21 | Pokhara |

### Unstructured Data
Unstructured data does not come in a neat table.

Examples:
- emails
- PDFs
- customer reviews
- audio
- video
- images
- social media posts

Example:
> "The delivery was late and the product box was damaged."

That sentence has useful information, but it is not in rows and columns.

### Semi-structured Data
This is in between.

Examples:
- JSON
- XML
- log files

These have some structure, but not as clean as a normal table.


## 4. Why We Need Data Cleaning

We clean data because:

1. wrong data gives wrong insights  
2. machine learning models need proper input  
3. dashboards and reports break with dirty data  
4. duplicate records can overcount things  
5. bad text formats reduce accuracy  
6. missing values can create bias or errors  

### Real-world examples
- In banking, wrong customer age can affect loan decisions
- In hospitals, missing patient records can affect treatment
- In e-commerce, duplicate orders can inflate revenue
- In stock analysis, wrong price type can distort indicators
- In education, wrong marks can lead to wrong student ranking

# 12. Cleaning Step by Step

Now we will clean the dataset properly.

We will:
1. copy the original data
2. clean text formatting
3. remove duplicates
4. fix gender values
5. standardize city names
6. fix age
7. fix salary
8. fix join date
9. handle missing values
10. inspect final cleaned dataset


```python
# Make a copy so the original raw data stays unchanged.
clean_df = df.copy()

# Show the copied version.
clean_df
```

## 12.1 Clean Column Names

Good column names are easier to work with.
We usually:
- convert to lowercase
- replace spaces with underscores


```python
# str.lower() converts column names to lowercase.
# str.strip() removes extra spaces from the start and end.
# str.replace(" ", "_") replaces spaces with underscores.
clean_df.columns = clean_df.columns.str.lower().str.strip().str.replace(" ", "_")

clean_df.columns
```

## 12.2 Clean Basic Text Formatting

We clean text columns by:
- removing extra spaces
- converting to consistent case


```python
# We select text columns manually.
text_columns = ["name", "gender", "city", "join_date"]

# Loop through each text column.
for col in text_columns:
    # astype("string") converts values to pandas string type.
    # str.strip() removes leading and trailing spaces.
    clean_df[col] = clean_df[col].astype("string").str.strip()

clean_df
```

## 12.3 Remove Duplicate Rows

We saw one duplicate row earlier.
Now we remove it.


```python
# drop_duplicates() removes exact duplicate rows.
clean_df = clean_df.drop_duplicates()

# Reset index to make row numbering clean again.
clean_df = clean_df.reset_index(drop=True)

clean_df
```

## 12.4 Standardize Gender Values

Current gender values are inconsistent:
- Male
- female
- M
- F
- male
- f

We map them into standard values: `Male` and `Female`.


```python
# First convert everything to lowercase for easier matching.
clean_df["gender"] = clean_df["gender"].str.lower()

# replace() changes old values into new values.
clean_df["gender"] = clean_df["gender"].replace({
    "m": "male",
    "male": "male",
    "f": "female",
    "female": "female"
})

# Convert into title case so output looks clean.
clean_df["gender"] = clean_df["gender"].str.title()

clean_df["gender"]
```

## 12.5 Standardize City Names

Cities are also inconsistent:
- Kathmandu
- ktm
- Pokhara
- pokhara
- Pokhara
- lalitpur with extra space

We standardize them.


```python
# Convert all city names to lowercase first.
clean_df["city"] = clean_df["city"].str.lower()

# Replace abbreviations and inconsistent entries with standard names.
clean_df["city"] = clean_df["city"].replace({
    "ktm": "kathmandu"
})

# Use title() to make city names look proper.
clean_df["city"] = clean_df["city"].str.title()

clean_df["city"]
```

## 12.6 Fix Age Column

Problems in age:
- numbers stored as strings
- `twenty` is text, not numeric
- missing value
- impossible values like `130` and `-5`

We first convert age to numeric.


```python
# to_numeric() tries to convert values into numbers.
# errors='coerce' means invalid values become NaN.
clean_df["age"] = pd.to_numeric(clean_df["age"], errors="coerce")

clean_df["age"]
```

```python
# Let's inspect age statistics after conversion.
clean_df["age"].describe()
```

```python
# Mark impossible ages as missing.
# Here we assume valid human age is between 0 and 100.
clean_df.loc[(clean_df["age"] < 0) | (clean_df["age"] > 100), "age"] = np.nan

clean_df["age"]
```

Now we fill missing age values.

A common method is the **median** because it is less affected by outliers than the mean.


```python
# median() calculates the middle value.
age_median = clean_df["age"].median()

# fillna() replaces missing values with the median.
clean_df["age"] = clean_df["age"].fillna(age_median)

clean_df["age"]
```

## 12.7 Fix Salary Column

Problems in salary:
- `60,000` has a comma
- `not available` is text
- one missing value
- one extreme outlier: `9999999`

Let's clean it.


```python
# Convert salary to string first so we can safely use string operations.
clean_df["salary"] = clean_df["salary"].astype("string")

# Remove commas from salary text.
clean_df["salary"] = clean_df["salary"].str.replace(",", "", regex=False)

# Convert cleaned salary text into numeric.
# Invalid values become NaN.
clean_df["salary"] = pd.to_numeric(clean_df["salary"], errors="coerce")

clean_df["salary"]
```

```python
# Inspect salary statistics.
clean_df["salary"].describe()
```

The salary `9999999` is too large compared to the rest.
This is likely an outlier or data entry error.

For this teaching example, we will mark salary above 200000 as missing.


```python
# Replace unrealistic salary values with NaN.
clean_df.loc[clean_df["salary"] > 200000, "salary"] = np.nan

clean_df["salary"]
```

```python
# Fill missing salary with the median salary.
salary_median = clean_df["salary"].median()
clean_df["salary"] = clean_df["salary"].fillna(salary_median)

clean_df["salary"]
```

## 12.8 Fix Join Date

Our join date column contains many formats:
- `2024-01-15`
- `15/02/2024`
- `March 5 2024`
- `2024/06/01`
- invalid date like `2024-13-01`

We convert it into proper datetime.


```python
# to_datetime() converts date text into real datetime values.
# errors='coerce' changes invalid dates into NaT (missing datetime).
clean_df["join_date"] = pd.to_datetime(clean_df["join_date"], errors="coerce")

clean_df["join_date"]
```

Now we fill missing join dates.

For this demo, we use forward fill, which means:
- take the previous valid date and use it for the missing row

This is not always the correct method in real work, but it is useful for learning.


```python
# ffill() means forward fill.
clean_df["join_date"] = clean_df["join_date"].ffill()

clean_df["join_date"]
```

## 12.9 Fill Missing Gender, City, and Name

For text/categorical columns, a common option is the **mode**.
Mode means the most frequent value.


```python
# Fill missing gender with the mode.
gender_mode = clean_df["gender"].mode()[0]
clean_df["gender"] = clean_df["gender"].fillna(gender_mode)

# Fill missing city with the mode.
city_mode = clean_df["city"].mode()[0]
clean_df["city"] = clean_df["city"].fillna(city_mode)

# Fill missing name with a placeholder.
clean_df["name"] = clean_df["name"].fillna("Unknown")

clean_df
```
