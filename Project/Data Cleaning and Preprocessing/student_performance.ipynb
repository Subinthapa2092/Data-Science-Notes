Droping Rows with Missing Values

```python
# Step 1 - drop only where Name is missing (identity, can't impute)
df = df.dropna(subset=['Name'])
print("after dropna Name:", df.shape)   # 966
```

3) Fill Missing Values::

```python
## It fill with Median of the Final Grade
df['FinalGrade'] = df['FinalGrade'].fillna(df['FinalGrade'].median())
## Numeric to fill with mean
df['StudyHoursPerWeek'] = df['StudyHoursPerWeek'].fillna(df['StudyHoursPerWeek'].mean())
## For the Categorical Data Filling with (most Frequent Repeated Values:: Mode)
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
### Filling the Parental support with the Mode Value
df['ParentalSupport'] = df['ParentalSupport'].fillna(df['ParentalSupport'].mode()[0])
## Fill all numeric columns at once
num_cols = df.select_dtypes(include='number').columns
df[num_cols] = df[num_cols].fillna(df[num_cols].mean())
## Checking Missing values is there or not
df.isnull().sum()
```

4) Checking Data types and Changing it's

```python
print(df.dtypes) ## it show the Column data type
print(df.info())## it show the data type with the non null count together
```

```python
### Converting colunms different data type
### Converting the float student Id data type to the int
df['StudentID'] = df['StudentID'].astype(int)
## converting Online Classes taken object to Boolean
df['Online Classes Taken'] = df['Online Classes Taken'].astype(bool)
#### Converting the Object of the Gender to Category::
df['Gender'] = df['Gender'].astype('category')
### converting the extracurricular activities to int
df['ExtracurricularActivities'] = df['ExtracurricularActivities'].astype(int)
### Converting Parental Support to category::
df['ParentalSupport'] = df['ParentalSupport'].astype('category')
#### Converting the String number to actual number of the final grade
df['FinalGrade'] = pd.to_numeric(df['FinalGrade'], errors='coerce')
### errors = corerce turns invalid Values into Nan instead of crashing
df
```

4) Finding Duplicates and removing duplicates

```python
### Couting Duplicate rows ::
print(df.duplicated().sum())
### View the Duplicated Rows:: with keep method 'last' and 'fast'
print(df[df.duplicated(keep = 'last')])
## Check Duplicates in one column::
print(df['StudentID'].duplicated().sum())
```

```python
### Removing the Duplicated Items::
df = df.drop_duplicates() ## it drop all exact duplicate rows
df
```

```python
df.describe()## it show the statical Description
```

5) Doing Filters Outliers::

```python
## Removing rows where Attendance > 100
df = df[df['Attendance (%)'] <= 100]
## Removing rows where Study Hours is negative
df = df[df['Study Hours'] >= 0]
### Keep grades between 0 and 100
df = df[(df['FinalGrade'] >= 0) & (df['FinalGrade'] <= 100)]
```

```python
df = df.drop(columns=['Study Hours', 'Attendance (%)'])
df = df.copy()
```
